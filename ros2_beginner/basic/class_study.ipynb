```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 6, 0.01)

y = np.sin(np.pi * t)

plt.figure(figsize= (12,6))
plt.plot(t,y)
plt.grid()
```

```python
t1 = np.arange(0, 3, 0.01)
y1 = np.sin(np.pi * 5 * t1)

plt.figure(figsize= (12,6))
plt.plot(t1, y1)
plt.grid()
plt.show()
```

```python
def draw_sin(t, A, f, b):
    y = A * np.sin(np.pi * f * t) + b

    plt.figure(figsize = (12,6))
    plt.plot(t,y)
    plt.grid()
    plt.show()

t = np.arange(0, 3, 0.01)
draw_sin(t, 1, 1, 2)
```

```python
class DrawSin:
    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time

    def calc_sin(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.sin(2*np.pi * self.freq * self.t) + self.bias

    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize = (12,6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show()

tmp = DrawSin(1, 1, 0, 3)

print(tmp.__dict__)

ds = DrawSin(1, 1, 0, 3)
ds.draw_sin()

```

```python
ds.freq= 10
ds.draw_sin()
```

```python
class DrawSinusoidal(DrawSin):
    def calc_cos(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.cos(2*np.pi * self.freq * self.t) + self.bias

    def draw_cos(self):
        y = self.calc_cos()
        plt.figure(figsize = (12,6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show()

dc = DrawSinusoidal(1,1,0,3)
print(dc.__dict__)

dc.draw_cos()
```

```python
class DrawSinusoidal2(DrawSinusoidal):
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize = (12,6))
        plt.plot(self.t, y)
        plt.title('Sin Graph')
        plt.ylabel('sin')
        plt.xlabel('time')
        plt.grid()
        plt.show()

dc2 =DrawSinusoidal2(1,1,0,3)
dc2.draw_sin()
```

```python
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        super().__init__(amp, freq, bias, end_time)
        self.ts = ts

    def calc_sin(self):
        self.t = np.arange(0, self.end_time, self.ts)
        return self.amp * np.sin(2*np.pi * self.freq * self.t) + self.bias

ds3 = DrawSinusoidal3(1,1,0,3, 0.1)
print(ds3.__dict__)

ds3.draw_sin()
```
